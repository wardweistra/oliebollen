```python
from lxml import etree, html
import pandas as pd
import email

mhtmlfile = "www.flexiwebs2.nl.mhtml"
htmlfile = "html_results_2017.html"
csvfile = "oliebollentest_2017.csv"
```

```python
f = open(mhtmlfile, 'r')

p = email.parser.Parser()
msg = p.parse(f)

f.close()

for part in msg.walk():
    if part.get_content_maintype()=="multipart":
        continue
    if part.get_content_type()=="text/html":
        f = open(htmlfile,"wb")
        f.write(part.get_payload(decode=1))
        f.close()
```

```python
with open(htmlfile, "r") as f:
    page = f.read()
tree = html.fromstring(page)
```

```python
item_divs_pattern = etree.XPath("//div[contains(@class, 'easy-col-sm-12') and contains(@class, 'easy_block')]")
item_divs = item_divs_pattern(tree)
len(item_divs)
```

```python
html.tostring(item_divs[0])
```

```python
entry_list = []

for item_div in item_divs:
    data = {}
    
    data['Naam'] = item_div.xpath(".//div[contains(@class, 'easy-col-sm-12') and contains(@class, 'title')]/h1")[0].text.strip()
    data['Plaats op ranglijst'] = item_div.xpath(".//div[contains(@class, 'easy-col-sm-12') and contains(@class, 'title')]/span[@class='nr']")[0].text.strip()[1:]
    
    address = item_div.xpath(".//div[contains(@class, 'subcontent')]/div/b[text()='Locatie:']")[0].tail.split(' - ')
    if len(address) > 2:
        print("Warning")
    data['Adres'] = address[0].strip()
    data['Plaats'] = address[-1].strip()
    
    data['Conclusie'] = item_div.xpath(".//div[contains(@class, 'subcontent')]/div/b[text()='Conclusie:']")[0].tail
    data['Zelf aangemeld voor deelname'] = item_div.xpath(".//div[contains(@class, 'subcontent')]/div/b[text()='Zelf aangemeld voor deelname:']")[0].tail
    
    data['Prijs per stuk in euro'] = item_div.xpath(".//div[contains(@class, 'subcontent')]/div/div/b[text()='Prijs per stuk in euro:']")[0].tail
    data['Aanbieding'] = item_div.xpath(".//div[contains(@class, 'subcontent')]/div/div/b[text()='Aanbieding:']")[0].tail
    data['Gewicht in gram'] = item_div.xpath(".//div[contains(@class, 'subcontent')]/div/div/b[text()='Gewicht in gram:']")[0].tail
    data['Vetpercentage'] = item_div.xpath(".//div[contains(@class, 'subcontent')]/div/div/b[text()='Vetpercentage:']")[0].tail
    
    
    entry_list.append(data)
entry_dataframe = pd.DataFrame(entry_list)
entry_dataframe
```

```python
entry_dataframe['Prijs per stuk in euro'] = entry_dataframe['Prijs per stuk in euro'].str.replace('€', '')
entry_dataframe['Prijs per stuk in euro'] = entry_dataframe['Prijs per stuk in euro'].str.replace(',', '.')
entry_dataframe['Prijs per stuk in euro'] = entry_dataframe['Prijs per stuk in euro'].str.strip()
```

```python
entry_dataframe['Vetpercentage'] = entry_dataframe['Vetpercentage'].str.replace(',', '.')
```

```python
entry_dataframe.to_csv(csvfile, index=False)
```
