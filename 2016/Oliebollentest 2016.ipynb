# Oliebollentest 2016


https://www.ad.nl/redactie/oliebollentest/2016/lijst.html

```python
from lxml import etree, html
import requests
import pandas as pd

htmlfile = "AD.nl, het laatste nieuws uit binnen- en buitenland, sport en show _ AD.nl.html"
csvfile = "oliebollentest_2016.csv"
```

# Data extraction

```python
with open(htmlfile, "r") as f:
    page = f.read()
tree = html.fromstring(page)
```

```python
item_divs_pattern = etree.XPath("//div[contains(@class, 'item') and contains(@class, 'row') and not(contains(@id, 'prop'))]")
item_divs = item_divs_pattern(tree)
len(item_divs)
```

```python
entry_list = []

for item_div in item_divs:
    data = {}

    header = item_div.find("div/h2").text
    data['Naam'] = ' '.join(header.split(' ')[1:])
    
    positie = header.split(' ')[0][1:]
    if '*' in positie:
        data['Nevenvestiging'] = True
    else:
        data['Nevenvestiging'] = False
    data['Plaats op ranglijst'] = positie.replace("*", "")
    
    address = item_div.find("div/h4").text.split('-')
    
    data['Adres'] = address[0].strip()
    data['Plaats'] = address[1].strip()
    
    data['Eindcijfer'] = item_div.xpath(".//div/p/strong[text() = 'Eindcijfer:']")[0].getnext().text
    data['Conclusie'] = item_div.xpath(".//div/p/strong[text() = 'Conclusie:']")[0].tail

    table_div = item_div.getnext().getnext()
    data['Categorie'] = table_div.xpath(".//td[text() = 'categorie']")[0].getnext().text
    data['Prijs per stuk in €'] = table_div.xpath(".//td[text() = 'Prijs per stuk in €']")[0].getnext().text

    try:
        data['Aanbieding in €'] = table_div.xpath(".//td[text() = 'Aanbieding in €']")[0].getnext().text
    except:
        pass
    
    data['Gewicht (gram)'] = table_div.xpath(".//td[text() = 'Gewicht (gram)']")[0].getnext().text
    data['Vet in %'] = table_div.xpath(".//td[text() = 'Vet in %']")[0].getnext().text
    
    try:
        data['Opmerking'] = table_div.xpath(".//td[text() = 'Opmerking']")[0].getnext().text
    except:
        pass 
    
    entry_list.append(data)
entry_dataframe = pd.DataFrame(entry_list)
entry_dataframe
```
