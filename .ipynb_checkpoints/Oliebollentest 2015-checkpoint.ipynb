Exported the KML file from there

```python
from fastkml import kml
from lxml import etree
import pandas as pd
```

```python
kml_file = 'AD Oliebollentest 2015.kml'
```

```python
with open(kml_file, 'rt', encoding="utf-8") as myfile:
    doc=myfile.read()
```

```python
root = etree.parse(kml_file)
stringkml = etree.tostring(root)
k = kml.KML()
k.from_string(stringkml)
```

```python
features = list(k.features())
list_of_features = list(features[0].features())
len(list_of_features)
```

```python
entry_list = []
for feature in list_of_features:
    data = {}
    data['name'] = feature.name
    data['address'] = feature.address
    data['description'] = feature.description
    data['styleUrl'] = feature.styleUrl
    for element in feature.extended_data.elements:
        data[str(element.name)] = str(element.value)
    entry_list.append(data)
entry_dataframe = pd.DataFrame(entry_list)
entry_dataframe
```

```python
out_file = 'oliebollentest_2015.csv'
entry_dataframe.to_csv(out_file)
```
