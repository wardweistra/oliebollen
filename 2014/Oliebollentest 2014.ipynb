```python
from lxml import etree, html
import requests
import pandas as pd
import re
import json
import html as htmlparser

htmlfile = "LocalFocus infographic.html"
csvfile = "oliebollentest_2014.csv"
```

```python
with open(htmlfile, "r") as f:
    page = f.read()
tree = html.fromstring(page)
```

```python
item_div = tree.xpath("//script[contains(.,'themodel')]/text()")[0]
```

```python
datastring = re.search("var themodel = {(.*?)};", item_div).group(1)
datastring = "{"+datastring+"}"
print(datastring)
```

```python
data = json.loads(datastring)
entry_list = []

positie = 1

for vendor in data['data']:
    data = {}
    
    namecity = vendor['label'].split('(')
    data['Naam'] = htmlparser.unescape(namecity[0].strip())
    data['Plaats'] = namecity[1][:-1].strip()
    data['Plaats op ranglijst'] = positie
    
    for value in vendor['element']['values']:
        data[value['lffilter1']] = value['value']
        
    entry_list.append(data)
    positie += 1
entry_dataframe = pd.DataFrame(entry_list)
entry_dataframe
```

```python
list(entry_dataframe)
```

```python
entry_dataframe.to_csv(csvfile, index=False)
```
